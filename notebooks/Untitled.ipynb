```python
from pathlib import Path

DATA_ROOT = Path("../data/rfuav_subset")
TRAIN_DIR = DATA_ROOT / "ImageSet-AllDrones-MatlabPipeline" / "train"

if TRAIN_DIR.exists():
    classes = sorted([d.name for d in TRAIN_DIR.iterdir() if d.is_dir()])
    print(f"Знайдено {len(classes)} класів:")
    for c in classes:
        n_files = len(list((TRAIN_DIR / c).glob("*.jpg")))
        print(f"  {c}: {n_files} зображень")
else:
    print(f"Папка {TRAIN_DIR} ще не існує — почекай завершення завантаження")
```
